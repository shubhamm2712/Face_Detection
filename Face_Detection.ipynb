```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=plt.imread("group_photo.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
detector
```

```python
faces=detector.detectMultiScale(img)
```

```python
faces[0]
```

```python
x,y,w,h=faces[0]
```

```python
modified_img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
plt.imshow(modified_img)
```

```python
modi_img=img
for face in faces:
    x,y,w,h=face
    modi_img=cv2.rectangle(modi_img,(x,y),(x+w,y+h),(255,0,0),5)
```

```python
plt.imshow(modi_img)
```
